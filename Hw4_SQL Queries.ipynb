```python
SQL:

SELECT hour,
       count(*)
FROM
  (SELECT extract(hour
                  FROM rental_date) AS hour
   FROM rental) AS hour
GROUP BY hour
ORDER BY COUNT DESC
```

```python
Query 2:
Make a list of all customer names who have videos out and how much they owe
```

```python
SQL:

SELECT c.first_name,
       c.last_name,
       sum(p.amount)
FROM customer c,
  (SELECT r.customer_id AS customer_id,
          r.rental_id AS rental_id
   FROM rental r
   WHERE r.return_date IS NULL) AS N,
     payment p
WHERE c.customer_id = N.customer_id
  AND p.rental_id = N.rental_id
GROUP BY c.first_name, c.last_name
ORDER BY sum DESC
```

```python
Query 3
Give the most popular actors by store location:
```

```python
SQL:

SELECT distinct add.address,COUNT(f.actor_id), first_name, last_name
FROM rental r
JOIN inventory i on i.inventory_id = r.inventory_id
JOIN store s on s.store_id = i.store_id
JOIN address add on add.address_id = s.address_id
JOIN film_actor f on f.film_id = i.film_id
JOIN actor act on act.actor_id = f.actor_id
GROUP BY add.address,first_name,last_name
ORDER BY COUNT DESC LIMIT 4
```
